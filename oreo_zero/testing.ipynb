```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import chess
```

```python
model = keras.models.load_model('oreo_zero_v2.keras')
```

```python
test_positions = np.load('test_data/test_positions.npy')
test_move_probs = np.load('test_data/test_moveProbs.npy')
test_outcomes = np.load('test_data/test_outcomes.npy')
```

```python
policy_predictions = []
value_predictions = []
for position in test_positions:
    network_output = model.predict(np.array([position]))
    policy_predictions.append(network_output[0][0])
    value_predictions.append(network_output[1][0][0])
```

```python
policy_predictions = np.array(policy_predictions)
value_predictions = np.array(value_predictions)

# Compute categorical crossentropy loss
cce = tf.keras.losses.CategoricalCrossentropy()
policy_loss = cce(test_move_probs, policy_predictions).numpy()

print("Categorical Crossentropy Loss:", policy_loss)
```
